```python
cleaned_sheets = []

for sheet_name in xl.sheet_names:
    
    sheet = xl.parse(sheet_name, header=None)
    
    header_row = None
    for i in range(len(sheet)):
        row = sheet.iloc[i]
        if "Date" in list(row):
            header_row = i
            break

    header = sheet.iloc[header_row]
    date_col = None
    for i in range(len(header)):
        if header[i] == "Date":
            date_col = i
            break

    data = sheet.iloc[header_row + 1:, date_col:date_col + 5]
    data.columns = ["Date", "Breakfast", "Lunch", "Evening_Snacks", "Dinner"]
    data["Date"] = pd.to_datetime(data["Date"], format="%d.%m.%Y", errors="coerce")
    data = data.dropna(subset=["Date"])

    cleaned_sheets.append(data)

print(len(cleaned_sheets))
```

```python
full_data = pd.concat(cleaned_sheets, ignore_index=True)
full_data = full_data.sort_values("Date").reset_index(drop=True)
full_data
```

```python
print(full_data["Date"].min(), full_data["Date"].max())
print(full_data["Date"].duplicated().sum())
print(full_data.isna().sum())
```
