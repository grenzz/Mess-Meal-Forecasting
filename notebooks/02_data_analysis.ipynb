```python
import pandas as pd
import matplotlib.pyplot as plt

training_data = pd.read_csv("/home/grenzz/funnnn/Mess-Meal-Forecasting/data/processed/mess_data_2024_2025_clean.csv", parse_dates=["Date"])
training_data.head()
```

```python
training_data.plot(x="Date", y="Lunch", figsize=(15,4), title="Lunch attendance over 2024-2025")
plt.show()
```

```python
subset = training_data[(training_data["Date"] >= "2024-08-01") & (training_data["Date"] <= "2024-11-01")]
subset.plot(x="Date", y="Lunch", figsize=(15,4), title="Lunch, Aug-Oct 2024")
plt.show()
```

```python
training_data.plot(x="Date", y=["Breakfast", "Lunch", "Evening_Snacks", "Dinner"], figsize=(15,5), title="All meals, 2024-2025")
plt.show()
```
